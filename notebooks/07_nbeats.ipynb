```python
# ============================================================
# Notebook 07 — N-BEATS Modelling
# Baseline V1: Interpretable stack, default configuration
# Reference: Oreshkin et al. (2020) ICLR
# Implementation: Nixtla neuralforecast library
# Author: Nihaar Raut | 001309432
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from neuralforecast import NeuralForecast
from neuralforecast.models import NBEATS
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')

PROCESSED = Path('../data/processed')
FIGURES = Path('../outputs/figures')

# Load all four series
ndvi_amazon_train = pd.read_csv(PROCESSED / 'ndvi_amazon_train.csv', parse_dates=['date'])
ndvi_amazon_test = pd.read_csv(PROCESSED / 'ndvi_amazon_test.csv', parse_dates=['date'])

ndvi_sahel_train = pd.read_csv(PROCESSED / 'ndvi_sahel_train.csv', parse_dates=['date'])
ndvi_sahel_test = pd.read_csv(PROCESSED / 'ndvi_sahel_test.csv', parse_dates=['date'])

sst_atlantic_train = pd.read_csv(PROCESSED / 'sst_atlantic_train.csv', parse_dates=['date'])
sst_atlantic_test = pd.read_csv(PROCESSED / 'sst_atlantic_test.csv', parse_dates=['date'])

sst_indian_train = pd.read_csv(PROCESSED / 'sst_indian_train.csv', parse_dates=['date'])
sst_indian_test = pd.read_csv(PROCESSED / 'sst_indian_test.csv', parse_dates=['date'])

print("All series loaded successfully")
```

```python
# ============================================================
# N-BEATS V1 — Baseline, interpretable stack
# Input size: 24 months (2 years lookback)
# Forecast horizon: 48 months (test period)
# Interpretable stack: trend + seasonality basis expansion
# neuralforecast requires specific dataframe format
# ============================================================

def prepare_nbeats_df(train, value_col, series_id):
    df = train[['date', value_col]].copy()
    df.columns = ['ds', 'y']
    df['unique_id'] = series_id
    df['ds'] = pd.to_datetime(df['ds'])
    return df

def run_nbeats(train, test, value_col, series_name, series_id):
    # Prepare data
    train_df = prepare_nbeats_df(train, value_col, series_id)
    
    # Define model
    model = NBEATS(
    h=48,
    input_size=24,
    stack_types=['trend', 'seasonality'],
    n_blocks=[3, 3],
    mlp_units=[[256, 256], [256, 256]],
    n_harmonics=2,
    n_polynomials=2,
    max_steps=500,
    learning_rate=1e-3,
    random_seed=42
    )
    
    nf = NeuralForecast(models=[model], freq='MS')
    nf.fit(train_df)
    
    # Forecast
    forecast_df = nf.predict()
    forecast = forecast_df['NBEATS'].values
    actual = test[value_col].values[:len(forecast)]
    
    # Metrics
    mae = mean_absolute_error(actual, forecast)
    rmse = np.sqrt(mean_squared_error(actual, forecast))
    
    # Plot
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(train['date'], train[value_col],
            color='#2c3e50', linewidth=1, label='Train')
    ax.plot(test['date'], actual,
            color='#27ae60', linewidth=1.5, label='Actual')
    
    forecast_dates = pd.date_range(
        start=test['date'].iloc[0], periods=len(forecast), freq='MS')
    ax.plot(forecast_dates, forecast,
            color='#c0392b', linewidth=1.5, linestyle='--', label='N-BEATS Forecast')
    ax.axvline(x=test['date'].iloc[0], color='grey',
               linestyle=':', linewidth=1, label='Train/Test Split')
    ax.set_title(f'N-BEATS V1 — {series_name} | MAE: {mae:.4f} | RMSE: {rmse:.4f}',
                 fontweight='bold')
    ax.set_xlabel('Date')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    
    filename = f"fig_nbeats_v1_{series_name.lower().replace(' ', '_')}.png"
    plt.savefig(FIGURES / filename, dpi=150, bbox_inches='tight')
    plt.show()
    
    return {'series': series_name, 'MAE': round(mae, 4), 'RMSE': round(rmse, 4)}

# Run V1 on all four series
results = []

print("Training N-BEATS — this will take several minutes...")
print()

results.append(run_nbeats(ndvi_amazon_train, ndvi_amazon_test,
                           'ndvi_mean', 'NDVI Amazon', 'ndvi_amazon'))

results.append(run_nbeats(ndvi_sahel_train, ndvi_sahel_test,
                           'ndvi_mean', 'NDVI Sahel', 'ndvi_sahel'))

results.append(run_nbeats(sst_atlantic_train, sst_atlantic_test,
                           'sst_mean', 'SST North Atlantic', 'sst_atlantic'))

results.append(run_nbeats(sst_indian_train, sst_indian_test,
                           'sst_mean', 'SST Indian Ocean', 'sst_indian'))

# Results table
results_df = pd.DataFrame(results)
print("\n=== N-BEATS V1 Baseline Results ===")
print(results_df.to_string(index=False))
```

```python
# ============================================================
# N-BEATS V2 — Tuned architecture with interpretable stack
# Improvements over V1 (default NeuralForecast settings):
#   - Increased input_size from 24 to 36 (3x seasonal period)
#   - Increased n_blocks and mlp_units for richer representation
#   - Interpretable stack: trend + seasonality decomposition
#   - max_steps increased from 100 to 500 for better convergence
#   - random_seed fixed for reproducibility
# Reference: Oreshkin et al. (2020) ICLR, arxiv:1905.10437
# ============================================================

from neuralforecast import NeuralForecast
from neuralforecast.models import NBEATS
from neuralforecast.losses.pytorch import MAE as NBEATS_MAE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

def prepare_nbeats_df(train, value_col, freq='MS'):
    """Convert to NeuralForecast long format."""
    df = train[['date', value_col]].copy()
    df.columns = ['ds', 'y']
    df.insert(0, 'unique_id', 'series_1')
    return df

def run_nbeats_v2(train, test, value_col, series_name):
    print(f"\nFitting N-BEATS V2 for {series_name}...")
    
    train_df = prepare_nbeats_df(train, value_col)
    
    # N-BEATS V2 — interpretable stack, tuned architecture
    models = [NBEATS(
        h=len(test),                    # forecast horizon = test length
        input_size=36,                  # 3 years lookback (3x m=12)
        stack_types=['trend', 'seasonality'],  # interpretable decomposition
        n_blocks=[3, 3],               # 3 blocks per stack
        mlp_units=[[512, 512],
                   [512, 512]],        # wider MLP layers
        n_harmonics=2,                 # Fourier terms for seasonality
        n_polynomials=3,               # polynomial degree for trend
        max_steps=500,                 # more training steps vs V1
        learning_rate=1e-3,
        loss=NBEATS_MAE(),
        random_seed=42,
        enable_progress_bar=False
    )]
    
    nf = NeuralForecast(models=models, freq='MS')
    nf.fit(train_df)
    
    # Forecast
    future_df = pd.DataFrame({
        'unique_id': 'series_1',
        'ds': test['date'].values
    })
    
    forecast_df = nf.predict()
    forecast = forecast_df['NBEATS'].values
    actual = test[value_col].values
    
    # Metrics
    mae = mean_absolute_error(actual, forecast)
    rmse = np.sqrt(mean_squared_error(actual, forecast))
    
    # Plot
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(train['date'], train[value_col],
            color='#2c3e50', linewidth=1, label='Train')
    ax.plot(test['date'], actual,
            color='#27ae60', linewidth=1.5, label='Actual')
    ax.plot(test['date'], forecast,
            color='#9b59b6', linewidth=1.5, linestyle='--',
            label='N-BEATS V2 Forecast')
    ax.axvline(x=test['date'].iloc[0], color='grey',
               linestyle=':', linewidth=1, label='Train/Test Split')
    ax.set_title(f'N-BEATS V2 — {series_name} | MAE: {mae:.4f} | RMSE: {rmse:.4f}',
                 fontweight='bold')
    ax.set_xlabel('Date')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    
    filename = f"fig_nbeats_v2_{series_name.lower().replace(' ', '_')}.png"
    plt.savefig(FIGURES / filename, dpi=150, bbox_inches='tight')
    plt.show()
    
    return {
        'series': series_name,
        'input_size': 36,
        'max_steps': 500,
        'MAE': round(mae, 4),
        'RMSE': round(rmse, 4)
    }

# Run V2 on all four series
results_nbeats_v2 = []

results_nbeats_v2.append(run_nbeats_v2(
    ndvi_amazon_train, ndvi_amazon_test,
    'ndvi_mean', 'NDVI Amazon'))

results_nbeats_v2.append(run_nbeats_v2(
    ndvi_sahel_train, ndvi_sahel_test,
    'ndvi_mean', 'NDVI Sahel'))

results_nbeats_v2.append(run_nbeats_v2(
    sst_atlantic_train, sst_atlantic_test,
    'sst_mean', 'SST North Atlantic'))

results_nbeats_v2.append(run_nbeats_v2(
    sst_indian_train, sst_indian_test,
    'sst_mean', 'SST Indian Ocean'))

# Results table
results_nbeats_v2_df = pd.DataFrame(results_nbeats_v2)
print("\n=== N-BEATS V2 Results ===")
print(results_nbeats_v2_df.to_string(index=False))
```

```python
# ============================================================
# N-BEATS V3 — Identity/Generic stack, increased capacity
# Key change from V2 (interpretable stack):
#   - stack_types=['identity','identity','identity'] — learns freely
#   - n_blocks [3,3,3] — more blocks than V2
#   - max_steps 1000 vs 500 in V2
#   - input_size kept at 36 (3 years lookback)
# Rationale: interpretable stack constrains basis functions
# to polynomial trend + Fourier seasonality. Identity stack
# learns optimal basis freely — better for asymmetric or
# complex seasonal shapes (SST North Atlantic, NDVI Amazon)
# Reference: Oreshkin et al. (2020) ICLR arxiv:1905.10437
# ============================================================

from neuralforecast import NeuralForecast
from neuralforecast.models import NBEATS
from neuralforecast.losses.pytorch import MAE as NBEATS_MAE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

def prepare_nbeats_df(train, value_col, freq='MS'):
    df = train[['date', value_col]].copy()
    df.columns = ['ds', 'y']
    df.insert(0, 'unique_id', 'series_1')
    return df

def run_nbeats_v3(train, test, value_col, series_name):
    print(f"\nFitting N-BEATS V3 for {series_name}...")
    
    train_df = prepare_nbeats_df(train, value_col)
    
    models = [NBEATS(
        h=len(test),
        input_size=36,
        stack_types=['identity', 'identity', 'identity'],
        n_blocks=[3, 3, 3],
        mlp_units=[[512, 512], [512, 512], [512, 512]],
        max_steps=1000,
        learning_rate=1e-3,
        loss=NBEATS_MAE(),
        random_seed=42,
        enable_progress_bar=False
    )]
    
    nf = NeuralForecast(models=models, freq='MS')
    nf.fit(train_df)
    
    forecast_df = nf.predict()
    forecast = forecast_df['NBEATS'].values
    actual = test[value_col].values
    
    mae = mean_absolute_error(actual, forecast)
    rmse = np.sqrt(mean_squared_error(actual, forecast))
    
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(train['date'], train[value_col],
            color='#2c3e50', linewidth=1, label='Train')
    ax.plot(test['date'], actual,
            color='#27ae60', linewidth=1.5, label='Actual')
    ax.plot(test['date'], forecast,
            color='#9b59b6', linewidth=1.5, linestyle='--',
            label='N-BEATS V3 Forecast')
    ax.axvline(x=test['date'].iloc[0], color='grey',
               linestyle=':', linewidth=1, label='Train/Test Split')
    ax.set_title(f'N-BEATS V3 (Generic) — {series_name} | MAE: {mae:.4f} | RMSE: {rmse:.4f}',
                 fontweight='bold')
    ax.set_xlabel('Date')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    
    filename = f"fig_nbeats_v3_{series_name.lower().replace(' ', '_')}.png"
    plt.savefig(FIGURES / filename, dpi=150, bbox_inches='tight')
    plt.show()
    
    return {
        'series': series_name,
        'stack': 'identity x3',
        'input_size': 36,
        'max_steps': 1000,
        'MAE': round(mae, 4),
        'RMSE': round(rmse, 4)
    }

# Run V3 on all four series
results_nbeats_v3 = []

results_nbeats_v3.append(run_nbeats_v3(
    ndvi_amazon_train, ndvi_amazon_test,
    'ndvi_mean', 'NDVI Amazon'))

results_nbeats_v3.append(run_nbeats_v3(
    ndvi_sahel_train, ndvi_sahel_test,
    'ndvi_mean', 'NDVI Sahel'))

results_nbeats_v3.append(run_nbeats_v3(
    sst_atlantic_train, sst_atlantic_test,
    'sst_mean', 'SST North Atlantic'))

results_nbeats_v3.append(run_nbeats_v3(
    sst_indian_train, sst_indian_test,
    'sst_mean', 'SST Indian Ocean'))

results_nbeats_v3_df = pd.DataFrame(results_nbeats_v3)
print("\n=== N-BEATS V3 Results ===")
print(results_nbeats_v3_df.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import mean_absolute_error, mean_squared_error
from neuralforecast import NeuralForecast
from neuralforecast.models import NBEATS, NBEATSx
from neuralforecast.losses.pytorch import MAE as NBEATS_MAE

DATA = Path('../data/processed')
FIGURES = Path('../outputs/figures')

# Load all series
ndvi_amazon_train = pd.read_csv(DATA / 'ndvi_amazon_train.csv')
ndvi_amazon_test = pd.read_csv(DATA / 'ndvi_amazon_test.csv')
ndvi_sahel_train = pd.read_csv(DATA / 'ndvi_sahel_train.csv')
ndvi_sahel_test = pd.read_csv(DATA / 'ndvi_sahel_test.csv')
sst_atlantic_train = pd.read_csv(DATA / 'sst_atlantic_train.csv')
sst_atlantic_test = pd.read_csv(DATA / 'sst_atlantic_test.csv')
sst_indian_train = pd.read_csv(DATA / 'sst_indian_train.csv')
sst_indian_test = pd.read_csv(DATA / 'sst_indian_test.csv')
co2_train = pd.read_csv(DATA / 'co2_train.csv')
co2_test = pd.read_csv(DATA / 'co2_test.csv')

for df in [ndvi_amazon_train, ndvi_amazon_test, ndvi_sahel_train,
           ndvi_sahel_test, sst_atlantic_train, sst_atlantic_test,
           sst_indian_train, sst_indian_test, co2_train, co2_test]:
    df['date'] = pd.to_datetime(df['date'])

print("All loaded successfully")
```

```python
# ============================================================
# Phase B — NBEATSx + CO₂ as exogenous variable
# NBEATSx extends N-BEATS with exogenous input support
# CO₂ passed as hist_exog (historical exogenous variable)
# Comparison: N-BEATS V3 (no CO₂) vs NBEATSx + CO₂
# Reference: Olivares et al. (2023) IJF
#            Oreshkin et al. (2020) ICLR
# ============================================================

def prepare_nbeatsx_df(train, value_col, co2_df):
    """Convert to NeuralForecast long format with CO₂ exog."""
    df = train[['date', value_col]].copy()
    df.columns = ['ds', 'y']
    df.insert(0, 'unique_id', 'series_1')
    # Merge CO₂
    co2_aligned = co2_df[co2_df['date'].isin(train['date'])][['date', 'co2_mean']].copy()
    co2_aligned.columns = ['ds', 'co2_mean']
    df = df.merge(co2_aligned, on='ds', how='left')
    return df

def prepare_test_df(test, co2_test_df):
    """Test df with CO₂ for NBEATSx prediction."""
    df = test[['date']].copy()
    df.columns = ['ds']
    df.insert(0, 'unique_id', 'series_1')
    df = df.merge(co2_test_df.rename(columns={'date': 'ds'})[['ds', 'co2_mean']],
                  on='ds', how='left')
    return df

def run_nbeatsx_co2(train, test, value_col, series_name):
    print(f"\nFitting NBEATSx + CO₂ for {series_name}...")

    train_df = prepare_nbeatsx_df(train, value_col, co2_train)
    print(f"  Train df shape: {train_df.shape}, CO₂ nulls: {train_df['co2_mean'].isna().sum()}")

    models = [NBEATSx(
        h=len(test),
        input_size=36,
        hist_exog_list=['co2_mean'],  # CO₂ as historical exogenous
        stack_types=['identity', 'identity', 'identity'],
        n_blocks=[3, 3, 3],
        mlp_units=[[512, 512], [512, 512], [512, 512]],
        max_steps=1000,
        learning_rate=1e-3,
        loss=NBEATS_MAE(),
        random_seed=42,
        enable_progress_bar=False
    )]

    nf = NeuralForecast(models=models, freq='MS')
    nf.fit(train_df)

    forecast_df = nf.predict()
    forecast = forecast_df['NBEATSx'].values
    actual = test[value_col].values

    mae = mean_absolute_error(actual, forecast)
    rmse = np.sqrt(mean_squared_error(actual, forecast))

    # N-BEATS V3 baseline MAE for comparison
    nbeats_v3_mae = {
        'NDVI Amazon': 0.0106,
        'NDVI Sahel': 0.0104,
        'SST North Atlantic': 0.5154,
        'SST Indian Ocean': 0.2119
    }
    mae_base = nbeats_v3_mae[series_name]

    # Plot
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(train['date'], train[value_col],
            color='#2c3e50', linewidth=1, label='Train')
    ax.plot(test['date'], actual,
            color='#27ae60', linewidth=1.5, label='Actual')
    ax.plot(test['date'], forecast,
            color='#8e44ad', linewidth=1.5, linestyle='--',
            label=f'NBEATSx+CO₂ MAE:{mae:.4f}')
    ax.axvline(x=test['date'].iloc[0], color='grey',
               linestyle=':', linewidth=1, label='Train/Test Split')
    ax.set_title(f'NBEATSx+CO₂ — {series_name} | MAE: {mae:.4f} | RMSE: {rmse:.4f}',
                 fontweight='bold')
    ax.set_xlabel('Date')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()

    filename = f"fig_nbeatsx_co2_{series_name.lower().replace(' ', '_')}.png"
    plt.savefig(FIGURES / filename, dpi=150, bbox_inches='tight')
    plt.show()

    diff = ((mae - mae_base) / mae_base) * 100
    direction = "↓ better" if mae < mae_base else "↑ worse"

    return {
        'series': series_name,
        'NBEATS_V3_MAE': mae_base,
        'NBEATSx_CO2_MAE': round(mae, 4),
        'change_pct': round(diff, 1),
        'verdict': direction
    }

# Run on all four series
results_nbeatsx = []

results_nbeatsx.append(run_nbeatsx_co2(
    ndvi_amazon_train, ndvi_amazon_test,
    'ndvi_mean', 'NDVI Amazon'))

results_nbeatsx.append(run_nbeatsx_co2(
    ndvi_sahel_train, ndvi_sahel_test,
    'ndvi_mean', 'NDVI Sahel'))

results_nbeatsx.append(run_nbeatsx_co2(
    sst_atlantic_train, sst_atlantic_test,
    'sst_mean', 'SST North Atlantic'))

results_nbeatsx.append(run_nbeatsx_co2(
    sst_indian_train, sst_indian_test,
    'sst_mean', 'SST Indian Ocean'))

results_nbeatsx_df = pd.DataFrame(results_nbeatsx)
print("\n=== NBEATSx + CO₂ Results ===")
print(results_nbeatsx_df.to_string(index=False))
```
